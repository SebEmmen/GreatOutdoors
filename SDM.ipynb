# Lezen van data into SDM

## Tabel Lezen Voorbereiding

```python
from pathlib import Path
import sqlite3
import pandas as pd
import os

source_folder = Path("GreatOutdoors")


crm_db_file = 'CRM-data.sqlite'
go_sales_file = 'GO_SALES-data.sqlite'
go_staff_file = 'GO_STAFF-data.sqlite'

target_db = source_folder / "SDM.sqlite"


crm = {
    "age_group": "Age_Group",
    "crm_country": "CRM_Country",
    "customer": "Customer",
    "customer_contact": "Customer_Contact",
    "customer_headquarters": "Customer_Headquarters",
    "customer_segment": "Customer_Segment",
    "customer_store": "Customer_Store",
    "customer_type": "Customer_Type",
    "sales_demographic": "Sales_Demographic",
    "sales_territory": "Sales_Territory"
}

sales = {
    "country": "Sales_Country",
    "order_details": "Order_Details",
    "order_header": "Order_Header",
    "order_method": "Order_Method",
    "product": "Product",
    "product_line": "Product_Line",
    "product_type": "Product_Type",
    "retailer_site": "Retailer_Site",
    "return_reason": "Return_Reason",
    "returned_item": "Returned_Item",
    "sales_branch": "Sales_Branch",
    "sales_staff": "Sales_Staff"
}

staff = {
    "course": "Course",
    "sales_office": "Sales_Office",
    "sales_representative": "Sales_Representative",
    "satisfaction": "Satisfaction",
    "satisfaction_type": "Satisfaction_Type",
    "training": "Training"
}
```

## CRM file lezen

```python
def load_crm_db_file():
    # 1. Connect to Source
    src_path = os.path.join(source_folder, crm_db_file)
    if not os.path.exists(src_path):
        print(f"❌ Could not find {crm_db_file} at {src_path}")
        return

    src_conn = sqlite3.connect(src_path)
    # 2. Connect to Target (this creates the file if it doesn't exist)
    tgt_conn = sqlite3.connect(target_db)
    
    print(f"🚀 Reading tables from {crm_db_file}...")

    for src_name, tgt_name in crm.items():
        try:
            # Read the data from the source table
            df = pd.read_sql_query(f"SELECT * FROM {src_name}", src_conn)
            
            # Save to the target database with the NEW prefixed name
            # if_exists='replace' ensures a clean table is created
            df.to_sql(tgt_name, tgt_conn, if_exists='replace', index=False)
            
            print(f"   ✅ Moved [{src_name}] -> [{tgt_name}] ({len(df)} rows)")
            
        except Exception as e:
            print(f"   ⚠️ Error processing table {src_name}: {e}")

    src_conn.close()
    tgt_conn.close()
print("\n✨ Data successfully loaded into the target SDM.")
load_crm_db_file()
```
